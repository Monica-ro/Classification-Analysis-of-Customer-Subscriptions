```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(font_scale=1.5)
# Grouped boxplot
sns.boxplot(x='y', y='emp.var.rate', data=bank_df)

# Calculate statistics for labels
infos = bank_df.groupby('y')['emp.var.rate'].describe()

   
# Set plot title and labels
plt.title('employment variation rate')
plt.xlabel('y')
plt.ylabel('emp.var.rate')

# Show the plot
plt.show()
```

```python
sns.set(font_scale=1.2)
sns.boxplot(x='y', y='cons.price.idx', data=bank_df)

# Calculate statistics for labels
stats = bank_df.groupby('y')['cons.price.idx'].describe()

# Set plot title and labels
plt.title('consumer price index')
plt.xlabel('y')
plt.ylabel('cons.price.idx')

plt.show()
```

```python
sns.boxplot(x='y', y='cons.conf.idx', data=bank_df)

# Calculate statistics for conf
stats = bank_df.groupby('y')['cons.conf.idx'].describe()

# Set plot title and labels
plt.title('consumer confidence index')
plt.xlabel('y')
plt.ylabel('cons.conf.idx')

plt.show()
```

```python
sns.boxplot(x='y', y='euribor3m', data=bank_df)

# Calculate statistics for conf
stats = bank_df.groupby('y')['euribor3m'].describe()

# Set plot title and labels
plt.title('euribor 3 month rate')
plt.xlabel('y')
plt.ylabel('euribor3m')


plt.show()
```

```python
sns.boxplot(x='y', y='nr.employed', data=bank_df)

# Calculate statistics for conf
stats = bank_df.groupby('y')['nr.employed'].describe()

# Set plot title and labels
plt.title('quarterly average of the total number of employed citizens')
plt.xlabel('y')
plt.ylabel('nr.employed')

plt.show()
```

```python
sns.countplot(x='contact', hue='y', data=bank_df)

# Set plot title and labels
plt.title('Bar Chart of Contact Methods grouped by Subscription Status')
plt.xlabel('contact')
plt.ylabel('Count')

plt.show()
```

```python
sns.countplot(x='education', hue='y', data=bank_df)

# Set plot title and labels
plt.title('Education grouped by Subscription Status')
plt.xticks(rotation=45)

plt.ylabel('Count')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'bank_df' is the DataFrame containing the 'pdays' and 'education' columns
fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(15, 5))

# Density Curve
axes[0].set_title('Density Curve for Age')
bank_df['age'].plot.kde(ax=axes[0])
axes[0].set_xlabel('age')
axes[0].set_ylabel('Density')

# Bar Chart
axes[1].set_title('Job Distribution')
job_counts = bank_df['job'].value_counts()
job_counts.plot.bar(ax=axes[1])
axes[1].set_xlabel('job')
axes[1].set_ylabel('Count')

## marital
axes[2].set_title('martial Distribution')
marital_counts = bank_df['marital'].value_counts()
marital_counts.plot.bar(ax=axes[2])
axes[2].set_xlabel('marital')
axes[2].set_ylabel('Count')

# Education Chart
axes[3].set_title('Education Distribution')
education_counts = bank_df['education'].value_counts()
education_counts.plot.bar(ax=axes[3])
axes[3].set_xlabel('education')
axes[3].set_ylabel('Count')

# y Chart
axes[4].set_title('Y Distribution')
education_counts = bank_df['y'].value_counts()
education_counts.plot.bar(ax=axes[4])
axes[4].set_xlabel('y')
axes[4].set_ylabel('Count')


plt.tight_layout()
plt.show()
```
